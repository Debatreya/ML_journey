## Gathering Data Using Web Scraping

#### Importing Required Libraries

```python
import pandas as pd
import requests
import bs4 as BeautifulSoup
```

```python
# We have to pass a Header so that it can by pass robots.txt file
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36'
}
webpage = requests.get('https://www.shiksha.com/colleges-4', headers=headers).text

webpage # This will give us the HTML content of the page
```

### Creating an Object of Beautiful Soup

```python
soup = BeautifulSoup.BeautifulSoup(webpage, 'html.parser')
print(soup.prettify())
```

#### Example Taking out all h2 tags

```python
for i in soup.find_all('p'):
    print(i.text.strip())
```
