## Parameters in `read_csv`

### Sep Parameter
Mention the separator, default = `,`(Comma), 

```python
pd.read_csv('./Data/movie_titles_metadata.tsv', sep='\t')
```

### Names Parameter:
By Default first row is Header, But here I have to mention my own header

```python
pd.read_csv('./Data/movie_titles_metadata.tsv', sep='\t', names=['sno', 'name', 'release_year', 'rating', 'votes', 'genres'])
```

### Index_col Parameter
Mention the Index_col

```python
pd.read_csv('./Data/movie_titles_metadata.tsv', sep='\t', names=['sno', 'name', 'release_year', 'rating', 'votes', 'genres'], index_col='sno')
```

### Header Parameter
Mention any row as the Header, like `header=1` makes the first row as header

```python
pd.read_csv('./Data/test.csv', header=1) # skiprows=1
# If noticed Before and After using header, It skips everything before the header row
```

### usecols parameter

Specific the required cols of the data

```python
pd.read_csv('./Data/aug_train.csv', usecols=['enrollee_id', 'gender', 'education_level'])
```

### Squeeze parameter (now a Method)
If we need only 1 col then `squeeze=True` creates a series rather than a DataFrame.

**It is deprecated as a Parameter, now we use like this**
`df.squeeze()`

```python
pd.read_csv('./Data/aug_train.csv', usecols=['enrollee_id']).squeeze()
```

### Skiprows/nrows Parameter

`Skiprows=[a, b]` to skip some rows (read docs) to remove ath and bth row of the csv file (0-based indexing)

`nrows=n` to select n number of rows

```python
pd.read_csv('./Data/aug_train.csv', skiprows=[1, 2, 3, 4], nrows=6)
```

### Encoding Parameter

use when UnicodeDecodeError is obtained, that is the dataset is using some other encoding rather than UTF-8

```python
# Find the encoding format and pass it as parameter
pd.read_csv('./Data/zomato.csv', encoding='latin-1')
```

### on_bad_lines Parameter

`on_bad_lines=skip` used to skip bad lines, like has extra parameter etc.

```python
pd.read_csv('./Data/zomato.csv', on_bad_lines='skip', encoding='latin-1', nrows=100)
```

### dtypes Parameter

Mention the data type of a column.

It is a dictionary parameter

```python
pd.read_csv('./Data/aug_train.csv').info()
# See here target -> float
```

```python
pd.read_csv('./Data/aug_train.csv', dtype={'target':int}).info()
```

### Handling Dates

By Default Dates are parsed as Strings. we need to use `parse_dates=['col_name']` to parse it as a datatime datatype and apply date logic on them

```python
## initially 'date' was of time object {string for pandas}
pd.read_csv('./Data/ipl.csv', parse_dates=['date']).info()
```

### Convertors

It is used to pre process every data cell before creating data frame.

It is a dictionary dtype

```python
# Convert 'Kolkata Knight Riders' to KKR
def reRR(name):
    if name == "Kolkata Knight Riders":
        return "KKR"
    return name

pd.read_csv('./Data/ipl.csv', converters={'team1': reRR, 'team2': reRR}).head()
```

```python
# correct the row 'sno' in movies tsv

def correctSerialNo(sno):
    return sno[1:]

pd.read_csv('./Data/movie_titles_metadata.tsv', sep='\t', names=['sno', 'name', 'release_year', 'rating', 'votes', 'genres'], converters={'sno': correctSerialNo}, index_col='sno').head()
```

### na_values Parameter

`NaN` is good then we can use dropna() fillna() type functions,
but sometimes we get only `-` in place of NaN values.

`na_values=['Male']` for eg. will specify 'Male' as a NaN type value

```python
pd.read_csv('./Data/aug_train.csv', na_values=['Male']).fillna('Female').head()
```
