```python
sns.displot(df['Fare'].dropna(), bins=25, kde=True, color='black', stat='density', aspect=2)
```

### c. Boxplot

Gives 5 number summary. (Outliers, Min, Max, Median, Q1, Q3, Lower Bound and Upper Bound)

Help to detect Noise and Outlier.
Above Upperbound and Below Lower Bound We have our outliers

```python
# AGE BOX PLOT


# Calculate the statistics
age_desc = df['Age'].describe()
min_val = age_desc['min']
q1 = age_desc['25%']
median = age_desc['50%']
q3 = age_desc['75%']
max_val = age_desc['max']
lower_bound = q1 - 1.5 * (q3 - q1)
upper_bound = q3 + 1.5 * (q3 - q1)

# Print the statistics
print('Min:', min_val)
print('Q1:', q1)
print('Median:', median)
print('Q3:', q3)
print('Max:', max_val)
print('IQR:', q3 - q1)
print('Range: ', max_val - min_val)
print('Lower Bound:', lower_bound)
print('Upper Bound:', upper_bound)


sns.boxplot(x='Age', data=df, orient='h', color='blue')
# sns.stripplot(x='Age', data=df, orient='h', color='red', jitter=True, size=5)
```

```python
# Fare BOX PLOT


# Calculate the statistics
Fare_desc = df['Fare'].describe()
min_val = Fare_desc['min']
q1 = Fare_desc['25%']
median = Fare_desc['50%']
q3 = Fare_desc['75%']
max_val = Fare_desc['max']
lower_bound = q1 - 1.5 * (q3 - q1)
upper_bound = q3 + 1.5 * (q3 - q1)

# Print the statistics
print('Min:', min_val)
print('Q1:', q1)
print('Median:', median)
print('Q3:', q3)
print('Max:', max_val)
print('IQR:', q3 - q1)
print('Range: ', max_val - min_val)
print('Lower Bound:', lower_bound)
print('Upper Bound:', upper_bound)


sns.boxplot(x='Fare', data=df, orient='h', color='blue')
# sns.stripplot(x='Fare', data=df, orient='h', color='red', jitter=True, size=5)
```
