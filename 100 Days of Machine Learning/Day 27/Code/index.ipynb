# Encoding Nominal Categorical Data

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./Data/cars.csv')
df.sample(5)
```

Nominal Categorical Data: `brand`, `fuel`, `owner`

```python
df['brand'].value_counts()
```

We Will Apply **One Hot Encoding** on these columns. 
We will apply the 'Max Frequency based Dimensionality Reduction' For `brand`

### 1. One Hot Encoding using Pandas

```python
pd.get_dummies(df, columns=['fuel', 'owner'])

# This is normal OneHotEncoding so we can expect 12 columns
# It has multicolinarity problem
```

Solving Multicollinearity Problem
### 2. K-1 One hot Encoding using Pandas

```python
pd.get_dummies(df, columns=['fuel', 'owner'], drop_first=True)

# This time expecting 10 columns
```

### 3. One Hot Encoding using sklearn

Better it tracks what changes it had done

```python
df.head()
```

First Do Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:4], df.iloc[:,-1], test_size=0.2, random_state=0)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(dtype=np.int32)
# ohe = OneHotEncoder(sparse=False) # if you want to return dense array
# ohe = OneHotEncoder(drop='first') # if you want to drop first column
```

```python
# Takeout the categorical columns
# Fit the encoder

X_train_new = ohe.fit_transform(X_train[['fuel', 'owner']]).toarray()
X_test_new = ohe.fit_transform(X_test[['fuel', 'owner']]).toarray()
```

```python
# Append the X_train_new to X_train

npX_train = np.hstack((X_train[['brand', 'km_driven']].values, X_train_new))

npX_train
```

```python
#  CONVERT TO A DATAFRAME
dfX_train = pd.DataFrame(npX_train)
dfX_train.head()
```

### 4. One Hot Encoding with Top Categories

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon'))
```
