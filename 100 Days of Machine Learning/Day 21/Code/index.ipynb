### 1. Scatterplot (Numerical - Numerical)

#### Bivariate

```python
#Scatterplot| tips : `total_bill` vs `tip`

sns.scatterplot(x='total_bill', y='tip', data=tips)
```

#### Multivariate

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker', size='size')
```

### 2. Bar Plot (Numerical - Categorical)

#### Bivariate

```python
# Barplot | titanic : `Pclass` vs `Age`

sns.barplot(x='Pclass', y='Age', data=titanic)
```

```python
# Barplot | titanic : `Pclass` vs `Fare`

sns.barplot(x='Pclass', y='Fare', data=titanic)
```

#### Multivariate

```python
# sns.barplot(x='Pclass', y='Age', data=titanic, hue='Sex')
sns.barplot(x='Pclass', y='Fare', data=titanic, hue='Sex')
```

### 3. Box Plot (Numerical -Categorical)

#### Bivariate

```python
# Boxplot | titanic: `Sex` vs `Age`

sns.boxplot(x='Sex', y='Age', data=titanic)
```

#### Multivariate

```python
sns.boxplot(x='Sex', y='Age', data=titanic, hue='Survived')
```

### 5. HeatMap (Categorical - Categorical)

Visualize N x N comparision (in numbers default)

```python
print(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']), annot=True, fmt='d')
```

```python
# TO get in terms of Percantage

# titanic.groupby('Embarked')['Survived'].mean()*100
# titanic.groupby('Sex')['Survived'].mean()*100
# titanic.groupby('Pclass')['Survived'].mean()*100

(titanic.groupby('Embarked')['Survived'].mean()*100).plot(kind='bar')
# (titanic.groupby('Sex')['Survived'].mean()*100).plot(kind='bar')
# (titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

### 8. Lineplot (Numerical - Numerical)

Visualize a large number of scatter plots. Used for time based data

```python
# Create a year wise flight data

new = flights.groupby('year').sum(numeric_only=True).reset_index()
print(new)

sns.lineplot(x='year', y='passengers', data=new)
```

```python
# Pivot Table
print(flights.pivot_table(index='month', columns='year', values='passengers'))
flights.pivot_table(index='month', columns='year', values='passengers').plot()
```

```python
# Heatmap
sns.heatmap(flights.pivot_table(index='month', columns='year', values='passengers'))
```

```python
# Clustermap
sns.clustermap(flights.pivot_table(index='month', columns='year', values='passengers'))
```
