# Taking out all movies

#### Create a empty dataframe

```python
# Create a DataFrame to store all movies initially empty
AllMovies = pd.DataFrame()
```

#### Loop through all the pages [total 428 pages] -> Make Request to each page -> create DF for each page -> Append to AllMovies

```python
for i in range(1, 30):
    url = 'https://api.themoviedb.org/3/movie/top_rated?api_key=d743f6ea15607da7e7e7813ef11d4212&&language=en-US&page=' + str(i);
    movieData = requests.get(url).json()
    movieData = movieData['results']
    print('Page ' + str(i) + ' of 429')
    temp_df = pd.DataFrame(movieData)[['id', 'title', 'vote_average', 'release_date', 'overview', 'popularity','vote_average' , 'vote_count']]
    AllMovies = pd.concat([AllMovies, temp_df], ignore_index=True)

AllMovies
```
